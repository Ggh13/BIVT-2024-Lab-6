```python
#   ,     
#      .
#  ,     

d = {'A': ['a1', 'a2', 'a3'],
     'B': ['b1', 'b2', 'b3', 'b4']}

d_inverted = {}
for k in d.keys():
  for v in d[k]:
      d_inverted.update({v:k})

d_inverted
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
encoders = {}
binary_categorical = ['isAlive', 'isAliveMother', 'isAliveFather', 'isAliveHeir', 'boolDeadRelations']
for col in binary_categorical:
    if col in data.columns:
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col])
        encoders[col] = le


```

```python
non_numeric_columns = data.select_dtypes(include=['object', 'category']).columns
print(f"\n : {list(non_numeric_columns)}")

#      
for col in non_numeric_columns:
  le = LabelEncoder()
  data[col] = le.fit_transform(data[col])
  encoders[col] = le
```
